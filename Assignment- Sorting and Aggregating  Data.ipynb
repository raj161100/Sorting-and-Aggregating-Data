#### `READ THE BIG MART SALES DATA`

```python
big_mart_data = pd.read_csv('dataset/big_mart_sales.csv')
```

```python
big_mart_data.head()
```

#### Find out the number of unique `Item_Type` in each of the `Outlet_Identifier`

---

```python
# WRITE YOUR CODE HERE
grouped = big_mart_data.groupby(['Outlet_Identifier','Item_Type'])
grouped.first()
```

```python
dnry = {}
for i in grouped['Outlet_Identifier','Item_Type']:
    l = []
    if i[0][0] in dnry:
        dnry[i[0][0]].extend([i[0][1]])
    elif i[0][0] not in dnry:
        l.append(i[0][1])
        dnry[i[0][0]] = l
        l = []
dnry.items()
for i in dnry.items():
    print("\n For each "+i[0]+" We have {} Items ".format(len(i[1])))
    print(i[1])
```

#### Find out the average `Item_Outlet_Sales` for each category of `Outlet_Size`

---

```python
# WRITE YOUR CODE HERE
new = big_mart_data.groupby(['Outlet_Size','Item_Outlet_Sales'])
new.first()
```

```python
dnry2 = {}
l = []
for i in new['Outlet_Size','Item_Outlet_Sales']:    
    #print(i[0][1])
    l = []
    if i[0][0] in dnry2:
        dnry2[i[0][0]].extend([i[0][1]])
    elif i[0][0] not in dnry2:
        l.append(i[0][1])
        dnry2[i[0][0]] = l

for i in dnry2.items():
    print("For "+i[0]+",  Average Item_Outlet_Sales is = {}.".format(sum(i[1])/len(i[1])))
```
